```python
def _sqrtm(C):
    # Computing diagonalization
    evalues, evectors = jnp.linalg.eig(C)
    # Ensuring square root matrix exists
    sqrt_matrix = evectors @ jnp.diag(jnp.sqrt(evalues)) @ jnp.linalg.inv(evectors)
    return sqrt_matrix.
```

```python
def h_13(xy,c12):
    return jnp.linalg.norm(xy, ord=2)**2 - c12

def h_2(x,y,c3):
    return np.dot(x,y) - c3

def B_(A, l1, l2, l3):
    I = jnp.identity(A.shape[0])
    B_11 = A + l1*I
    B_22 = A + l2*I
    B_12 = l3*I
    B_21 = l3*I
    B = jnp.block([[B_11, B_12],[B_21, B_22]])
    return B

def T_(A, l1, l2, l3):
    I = jnp.identity(A.shape[0])
    T = (A + l1*I)@(A + l2*I) - l3**2*I  
    return T
```

```python
def f(X, A, b, d):
    return X[:,0].T@A@X[:,0] + X[:,1].T@A@X[:,1] + 2*b.T@X[:,0] + 2*d.T@X[:,1]
```

```python
def g(X, v):
    return np.array([v.T@X[:,0], v.T@X[:,1]])
def h(X, D, c1, c2, c3):
    return np.array([X[:,0].T@D@X[:,0] - c1, X[:,1].T@D@X[:,1] - c2, 2*(X[:,0].T@D@X[:,1] - c3)])
```

```python
def project(X1, C):
    C1 = X1.T@X1
    C1sqrt = sqrtm(C1)
    Csqrt = sqrtm(C)
    U,s,V = np.linalg.svd(Csqrt@C1sqrt)
    X = X1@np.linalg.inv(C1sqrt)@U@V.T@Csqrt
    
    return X
```
